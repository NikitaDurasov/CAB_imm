```python
def build_decision_function(coef, intercept):
    return lambda x: (- coef[0][0]*x - intercept[0])/coef[0][1]
def build_rescaled_decision_function(coef, intercept, std, mean):
    new_a = coef[0][0] * np.sqrt(std[1])
    new_b = coef[0][1] * np.sqrt(std[0])
    new_c = (intercept[0]*np.sqrt(std[0])*np.sqrt(std[1]) - coef[0][0]*np.sqrt(std[1])*mean[0] - 
             coef[0][1]*np.sqrt(std[0])*mean[1])
    return [[[new_a, new_b]],[new_c]]
```
